## Problem 2


Write a function that takes a two-dimensional array and a string as input and returns an array rotated 90 degrees counterclockwise if the string 'left' was passed, and clockwise if the string 'right' was passed.

Example for input: $\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$.\
If the string 'left' is passed, the function should return $\begin{bmatrix} 3 & 6 & 9 \\ 2 & 5 & 8 \\ 1 & 4 & 7 \end{bmatrix}$, and if the string 'right' is passed, the function should return $\begin{bmatrix} 7 & 4 & 1 \\ 8 & 5 & 2 \\ 9 & 6 & 3 \end{bmatrix}$.

Do not use numpy and nested loops. Use list comprehensions and built-in function `zip`.

````python
def rotate_matrix(matrix, text):
    if(text=='right'):
        return[col[::-1] for col in zip(*matrix)]
    if(text=='left'):
        return[col for col in zip(*matrix)][::-1]
    else:
        return(-1)

matrix=[[1,2,3],
        [4,5,6],
        [7,8,9]]
print('Поворот направо')
for row in rotate_matrix(matrix,'right'):
    print(row)

print('Поворот налево')
for row in rotate_matrix(matrix,'left'):
    print(row)
````

## Problem 4

### Implementing a Library Management System

#### Description

You are required to design and implement a system for managing books and users in a library. The system should allow for the management of books (adding, deleting, searching by various criteria) and users (registration, deletion, searching), as well as tracking the history of interactions between them (issuing and returning books).

#### Tasks

1. **`Book` Class**:
   - Attributes: title, author, year of publication, ISBN, number of copies.
   - Methods: constructor, methods to get information about the book, method to change the number of copies (when issuing and returning books).

2. **`User` Class**:
   - Attributes: user name, library card number, list of borrowed books.
   - Methods: constructor, methods for user registration, methods for adding and removing books from the borrowed list.

3. **`Library` Class**:
   - Attributes: list of books, list of users, transaction history (who, when, which book was borrowed and returned).
   - Methods: constructor, methods for adding and deleting books and users, methods for issuing and returning books, searching for books and users by various criteria, method to display the transaction history.

#### Assignment

1. Implement the `Book`, `User`, and `Library` classes with the specified attributes and methods.
2. Create several books and users, and add them to the library system.
3. Implement scenarios for issuing books to users and their return.
4. Display the transaction history to show how books were issued and returned.


````python
class Item:
    def __init__(self, name, writer, pub_year, code, quantity):
        self.name = name
        self.writer = writer
        self.pub_year = pub_year
        self.code = code
        self.quantity = quantity
        print(f"[+] Книга '{name}' добавлена в каталог")

    def get_name(self): return self.name
    def get_writer(self): return self.writer
    def get_year(self): return self.pub_year
    def get_code(self): return self.code
    def get_qty(self): return self.quantity

    def update_qty(self, action):
        if action == "out":
            if self.quantity > 0:
                self.quantity -= 1
                return True
            return False
        elif action == "in":
            self.quantity += 1
            return True
        else:
            raise Exception("Неверный тип операции")


class Reader:
    def __init__(self, fullname, ticket):
        self.fullname = fullname
        self.ticket = ticket
        self.taken_items = []
        print(f"[+] Читатель {fullname} зарегистрирован")

    def update_info(self, new_name=None, new_ticket=None):
        if new_name: self.fullname = new_name
        if new_ticket: self.ticket = new_ticket

    def take_item(self, item):
        if item.get_qty() > 0:
            self.taken_items.append(item)
            item.update_qty("out")
            return True
        return False

    def give_back(self, item):
        if item in self.taken_items:
            self.taken_items.remove(item)
            item.update_qty("in")
            return True
        return False


class LibrarySystem:
    def __init__(self):
        self.readers = []
        self.catalog = []
        self.events = []
        print("=== Библиотечная система инициализирована ===\n")

    def reg_reader(self, reader):
        for r in self.readers:
            if r.ticket == reader.ticket:
                print("-> Читатель с таким номером уже есть")
                return False
        self.readers.append(reader)
        return True

    def unreg_reader(self, reader):
        if reader not in self.readers:
            print("-> Читатель не найден")
            return False
        if reader.taken_items:
            print("-> Сначала нужно вернуть книги")
            return False
        self.readers.remove(reader)
        return True

    def add_to_catalog(self, item):
        for it in self.catalog:
            if it.code == item.code:
                print("-> Книга с таким ISBN уже есть")
                return False
        self.catalog.append(item)
        return True

    def remove_from_catalog(self, item):
        if item not in self.catalog:
            print("-> Книга не найдена")
            return False
        self.catalog.remove(item)
        return True

    def issue(self, reader, item):
        if reader.take_item(item):
            self.events.append(f"[ВЫДАЧА] {reader.fullname} взял '{item.name}'")
            return True
        return False

    def receive(self, reader, item):
        if item in reader.taken_items:
            if reader.give_back(item):
                self.events.append(f"[ВОЗВРАТ] {reader.fullname} вернул '{item.name}'")
                return True
        print("-> Этой книги у читателя нет")
        return False

    def search_items(self, title=None, author=None, year=None, isbn=None):
        results = self.catalog
        if title:
            results = [i for i in results if title.lower() in i.get_name().lower()]
        if author:
            results = [i for i in results if author.lower() in i.get_writer().lower()]
        if year:
            results = [i for i in results if year == i.get_year()]
        if isbn:
            results = [i for i in results if isbn == i.get_code()]

        print(f"\n--- Найдено книг: {len(results)} ---")
        for i in results:
            print(f"  • {i.name} | {i.writer} ({i.pub_year}) | {i.code} | в наличии: {i.quantity}")

    def search_readers(self, name=None, ticket=None):
        results = self.readers
        if name:
            results = [r for r in results if name.lower() in r.fullname.lower()]
        if ticket:
            results = [r for r in results if ticket == r.ticket]

        print(f"\n--- Найдено читателей: {len(results)} ---")
        for r in results:
            taken = [i.name for i in r.taken_items]
            print(f"  • {r.fullname} (билет №{r.ticket}) | взято: {taken if taken else 'нет'}")

    def show_events(self):
        print("\n=== ЖУРНАЛ СОБЫТИЙ ===")
        for ev in self.events:
            print(ev)


# ----- ТЕСТИРОВАНИЕ -----
lib = LibrarySystem()

b1 = Item("Приключения Тома Сойера", "Марк Твен", 1876, "978-5-389-17627-0", 1)
b2 = Item("Волшебник Изумрудного города", "Александр Волков", 1939, "978-0-00-444152-8", 3)
b3 = Item("Алиса в стране чудес", "Льюис Кэрролл", 1865, "978-5-389-00818-2", 2)

lib.add_to_catalog(b1)
lib.add_to_catalog(b2)
lib.add_to_catalog(b3)

r1 = Reader("Жарков Николай", 123321)
r2 = Reader("Жарков Николай Иванович", 456654)
r3 = Reader("Николай", 789987)

lib.reg_reader(r1)
lib.reg_reader(r2)
lib.reg_reader(r3)

lib.search_items(author="Кэрролл")
lib.search_readers(ticket=123321)

lib.issue(r2, b2)
lib.issue(r1, b2)
lib.receive(r1, b2)
lib.issue(r3, b1)
lib.receive(r3, b1)
lib.issue(r2,b1)
lib.receive(r2, b1)

lib.show_events()
````
